```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("financials.csv")
```

```python
df
```

```python
#Filling the missing values in Columns 'Total Liabilities','Cash Flow from Operating Activities' with Mean
df[['Total Assets','Total Liabilities','Cash Flow from Operating Activities']] = df[['Total Assets','Total Liabilities','Cash Flow from Operating Activities']].fillna(df[['Total Assets','Total Liabilities','Cash Flow from Operating Activities']].mean())
```

```python
df
```

```python
# calculating the yearly percentage changes for each financial metric
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df['Total Assets Change (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Total Liabilities Change (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow Operating Activities Change (%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
df
```

```python
#filling the NAN value from percentage change calculation
df.fillna(0, inplace=True)
```

```python
df
```

```python
# Groupings by company
by_company = df.groupby('Company').agg({
    'Total Revenue': ['mean', 'min', 'max','sum'],
    'Net Income': ['mean', 'min', 'max','sum'],
    'Total Assets': ['mean', 'min', 'max','sum'],
    'Total Liabilities': ['mean', 'min', 'max', 'sum'],
    'Cash Flow from Operating Activities': ['mean','min', 'max', 'sum'] 
})
by_company
```

```python
# Summarize findings for each company
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Total Assets Change (%)' : 'mean',
    'Total Liabilities Change (%)' : 'mean',
    'Cash Flow Operating Activities Change (%)' :'mean'    
})
print("Overall Growth/Fall rate for Apple, Microsoft and Tesla from Fiscal Year 2021 - 2023")
summary
```

```python
plt.figure(figsize=(10, 6))
for company, data in df.groupby('Company'):
    plt.plot(data['Year'], data['Net Income Growth (%)'], label=company)
plt.title('Net Income Growth (%) Over Time')
plt.xlabel('Year')
plt.ylabel('Net Income Growth (%)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for company, data in df.groupby('Company'):
    plt.plot(data['Year'], data['Revenue Growth (%)'], label=company)
plt.title('Revenue Growth (%) Over Time')
plt.xlabel('Year')
plt.ylabel('Revenue Growth (%)')
plt.legend()
plt.grid(True)
plt.show()
```
